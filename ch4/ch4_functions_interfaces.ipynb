```python
##INSTALL jupyturtle

## !pip install jupyturtle

import math
from jupyturtle import make_turtle,forward,left,right,penup,pendown


def jump(length):
   
    penup()
    forward(length)
    pendown()

##EXERCISE 1

def rectangle(wide, tall) :

    for n in range(1,5):
        if(n % 2 == 0):
            forward(tall)
            left(90)
        else :
            forward(wide)
            left(90)

make_turtle()
rectangle(80,40)

##EXERCISE 2

def rhombus(length, angle) :
     for n in range(1,5):
        if(n % 2 == 0):
            forward(length)
            left(180 - angle)
        else :
            forward(length)
            left(angle)

make_turtle()
rhombus(50,60)

##EXERCISE 3

def parallelogram(length1, length2, angle) :
    for n in range(1,5):
        if(n % 2 == 0):
            forward(length2)
            left(180 - angle)
        else :
            forward(length1)
            left(angle)
def rectangle(wide, tall):
    parallelogram(wide, tall,90)

def rhombus(length,angle):
    parallelogram(length,length,angle)

make_turtle(width=800)

jump(-140)
rectangle(100,50)

jump(140)
rhombus(80,60)

jump(140)
parallelogram(100, 70, 60)

#EXERCISE 4

##prerequisite
def arc(radius, angle):
    arc_length = 2 * math.pi * radius * angle / 360
    n = 30
    length = arc_length / n
    step_angle = angle / n
    polyline(n, length, step_angle)

def polyline(n, length, angle):
    for i in range(n):
        forward(length)
        left(angle)

def triangle(radius, angle):
    
    y = radius * math.sin(angle * math.pi / 180)

    right(angle)

    forward(radius)
    left(90 + angle)

    forward(2*y)
    left(90 + angle)

    forward(radius)
    left(180 - angle)

def draw_pie(segments, radius):

    angle = 360.0 / segments

    for i in range(segments):
        triangle(radius,  angle/2)
        left(angle)

turtle = make_turtle(width = 600,delay=0)
jump(-80)

radius = 60
draw_pie(7, radius)

jump(2 * radius)
draw_pie(8, radius)

jump(2 * radius)
draw_pie(9, radius)

#EXERCISE 5

def petal(arc_radius,  angle):
   
    for i in range(2):
        arc(arc_radius,  angle)
        left(180 - angle)

def flower(petal_number, arc_radius, angle):
    
    for i in range(petal_number):
        petal(arc_radius,  angle)
        left(360.0/petal_number)


make_turtle()

jump(-60)
n = 7
radius = 60
angle = 60
flower(n, radius, angle)

jump(120)
n = 9
radius = 40
angle = 85
flower(n, radius, angle)

```
